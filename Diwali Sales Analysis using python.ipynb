## Gender

```python
# A bar chart for Gender & it's count
ax=sns.countplot(x='Gender',data=df,hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
    plt.title ('No. of Gender')
```

```python
df.groupby(['Gender'],as_index=False) ['Amount'].sum().sort_values (by='Amount',ascending=False)
```

```python
# Amount vs Gender
sales_gen=df.groupby(['Gender'],as_index=False) ['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen,hue='Amount')
for bars in ax.containers:
    ax.bar_label(bars)
    plt.title ('Total Amount Spend by Gender')
```

### From the above graphs most of the buyer are Females and Even purchasing power of Females greater than men.

## Age

```python
# No.of age groups
ax=sns.countplot(x='Age Group',data=df,hue='Age Group')
for bars in ax.containers:
    ax.bar_label(bars)
    plt.title ('No.of age group')
```

```python
## No of orders gender wise according to Age group
ax=sns.countplot(x='Age Group',data=df,hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)

plt.title ('No of buyers vs Age groups vs Gender')
```

### from the above graps we can see that most of the buyers age between 26-35 and most of them Female.

## State

```python
# total number of orders from top 10 states 
sales_state=df.groupby(['State'], as_index=False) ['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.barplot(x='State',y='Orders',data=sales_state,hue='State')
for bars in ax.containers:
    ax.bar_label(bars)
    plt.title ('Number of Orders from Top 10 State')
```

```python
# Total amount of sales from Top 10 states
sales_state=df.groupby(['State'], as_index=False) ['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,7)})
sns.barplot(x='State',y='Amount',data=sales_state,hue='State')
plt.title ('Amount of sales from Top 10 state')
```

### From the above two State graphs we can see that no.of order & the amount of sales are primarily Uttar pradesh,Maharashtra and Karnataka respectively, at the both graphs in small difference i.e Kerala is at number 8 in the total number of orders but in the graphs of the total amount kerala is not there.this shows that Kerala is No of orders and that too in low-average price items and not expensive items.The Store earn better from Harayana and Bhihara instead of kerala.   

## Marital Status

```python
# No of order according to marital status
ax=sns.countplot(x = 'Marital_Status',data =df,hue='Marital_Status')
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('No of Purchases vs Marital Status')

```

```python
# Total amount of sales according to Marital_Status and Gender
sales_ms=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Marital_Status',y='Amount',hue='Gender',data=sales_ms)
plt.title('Total amount vs Marital status vs Gender')
```

### From the above graphs we can see that the most of the buyers are Married(Women) and they have high purchasing power.

## occuption

```python
# No.of orders according to Occuption
ax=sns.countplot(x='Occupation',data=df)

sns.set(rc={'figure.figsize':(10,7)})
for bars in ax.containers:
    ax.bar_label(bars)
plt.title ('No.of buyers vs Occupation')
```

```python
# Total amount of sales according to occupation
sales_occ=df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Occupation',y='Amount',hue='Occupation',data=sales_occ,palette='inferno')

sns.set (rc={'figure.figsize':(20,5)})
plt.title ('Total amount of sales acccording to occupation')
```

### From the above graphs we can see that most of the buyers are working in IT,Healthcare,Aviation sector then so on.

## Product Category

```python
# No.of sales according to Product Category
ax=sns.countplot(x='Product_Category',data=df,hue='Product_Category')

sns.set(rc={'figure.figsize':(30,6)})
for bars in ax.containers:
    ax.bar_label(bars)
    
plt.title('No of sales according to Product Category')
```

```python
### 
```

```python
# Total amount of sales according to Product Category
sales_pc = df.groupby(['Product_Category'],as_index=False) ['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.set({'figure.figsize':(30,5)})
sns.barplot(x='Product_Category',y='Amount',data=sales_pc,hue='Product_Category')

plt.title('Total amount of sales according to Product Category')
```

### From the above two Product_category graphs we can see that some difference in Orders vs Amount of sales top no of orders Respectively cloths & apparel,food,Electronics & Gadgets but Amount of sales respectively Food,clothing & apparel,Elctronics & gadgets.

```python
# Product id of Top 10 selling products
df.groupby('Product_ID') ['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')

plt.title('Product ID of Top 10 Selling Products')

```
